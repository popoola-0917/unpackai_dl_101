```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("https://raw.githubusercontent.com/unpackAI/unpackai/main/examples/datasets/online_shoppers_intention.csv")
df.head()
```

#### 3. Explore your dataset using Pandas Profiling

```python
profile = ProfileReport(df, title="Shoppers Intention EDA")
profile
```

Split the dataset to training and test sets

```python
data = df.sample(frac=0.95, random_state=786)
data_unseen = df.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

#### 4. Automated Machine Learning

```python
from pycaret.classification import *
```

```python
data_prep = setup(data, target = 'Revenue', session_id=123, fix_imbalance=True, feature_selection = True, remove_multicollinearity = True)
```

```python
best_model = compare_models()
```

```python
lightgbm = create_model('lightgbm')
```

```python
tuned_light = tune_model(lightgbm)
```

```python
print(tuned_light)
```

```python
evaluate_model(tuned_light)
```

```python
predict_model(tuned_light);
```

```python
final_light = finalize_model(tuned_light)
```

```python
predict_model(final_light);
```

```python
unseen_predictions = predict_model(final_light, data=data_unseen)
unseen_predictions.head()
```

```python
from pycaret.utils import check_metric
check_metric(unseen_predictions['Revenue'], unseen_predictions['Label']=='True', metric = 'Accuracy')
```

```python
save_model(final_light,'Final Light Model 7Dec2021')
```

```python
saved_final_light = load_model('/content/Final Light Model 7Dec2021')
```

```python
new_prediction = predict_model(saved_final_light, data=data_unseen)
```
